```python
# 2. PULIZIA NOMI COLONNE
print(f"\nüßπ PULIZIA NOMI COLONNE:")

# Pulizia generale nomi colonne
df_startup.columns = df_startup.columns.str.strip()  # Rimuovi spazi
print(f"‚úÖ Spazi rimossi dai nomi colonne")

# Mappa nomi colonne pi√π leggibili (mantenendo riferimenti alla legenda)
column_mapping = {
    'denominazione': 'ragione_sociale',
    'nome': 'nome',
    'tipo_societa_Copy': 'tipo_societa',
    'stato': 'stato',
    'nat.giuridica': 'natura_giuridica',
    'pv': 'provincia',
    'comune': 'comune',
    'data iscrizione alla sezione delle startup': 'data_iscrizione_startup',
    'data iscrizione al Registro Imprese': 'data_iscrizione_ri',
    'data inizio dell\'esercizio effettivo dell\'attivit√†': 'data_inizio_attivita',
    'ateco 2007': 'codice_ateco',
    'settore': 'settore',
    'attivit√†': 'descrizione_attivita',
    'classe di produzione ultimo anno (1)': 'classe_produzione',  # IMPORTANTE: dati reali fatturato
    'classe di addetti ultimo anno (2)': 'classe_addetti',       # IMPORTANTE: dati reali dipendenti
    'impresa a vocazione sociale (5)': 'vocazione_sociale',
    'impresa ad alto valore tecnologico in ambito energetico(4)': 'alto_valore_tech_energia',
    'classe di capitale (3)': 'classe_capitale',                 # IMPORTANTE: capitale sociale
    'sito internet': 'sito_web',
    '1¬∞ req.\\n(6)': 'requisito_1_rd',     # 15% costi per R&D
    '2¬∞ req.\\n(6)': 'requisito_2_team',   # Team qualificato
    '3¬∞ req.\\n(6)': 'requisito_3_brevetti', # Brevetti/software
    'data dichiarazione\\n(7)': 'data_dichiarazione',
    'prevalenza femminile (8)': 'prevalenza_femminile',
    'prevalenza giovanile (8)': 'prevalenza_giovanile',
    'prevalenza straniera (8)': 'prevalenza_straniera'
}

# Applica mapping solo per colonne esistenti
existing_mapping = {old: new for old, new in column_mapping.items() if old in df_startup.columns}
df_startup.rename(columns=existing_mapping, inplace=True)

print(f"üìã COLONNE RINOMINATE:")
for old, new in existing_mapping.items():
    print(f"  '{old}' ‚Üí '{new}'")

# 3. CONTROLLO QUALIT√Ä DATI
print(f"\nüîç CONTROLLO QUALIT√Ä DATI:")

# Valori mancanti
missing_data = df_startup.isnull().sum()
missing_percent = (missing_data / len(df_startup)) * 100

print(f"üìà VALORI MANCANTI:")
missing_summary = pd.DataFrame({
    'Colonna': missing_data.index,
    'Valori_Mancanti': missing_data.values,
    'Percentuale': missing_percent.values
}).sort_values('Percentuale', ascending=False)

# Mostra solo colonne con valori mancanti > 0
missing_cols = missing_summary[missing_summary['Valori_Mancanti'] > 0]
if len(missing_cols) > 0:
    print(missing_cols.head(10).to_string(index=False))
else:
    print("üéâ Nessun valore mancante!")

# 4. CONTROLLO DUPLICATI
print(f"\nüîÑ CONTROLLO DUPLICATI:")
duplicates = df_startup.duplicated(subset=['ragione_sociale', 'provincia', 'data_iscrizione_startup']).sum()
print(f"Possibili duplicati: {duplicates}")

if duplicates > 0:
    print(f"‚ö†Ô∏è Trovati {duplicates} possibili duplicati")
    # Mostra esempio duplicati
    duplicate_rows = df_startup[df_startup.duplicated(subset=['ragione_sociale', 'provincia', 'data_iscrizione_startup'], keep=False)]
    print(f"Esempi duplicati:")
    print(duplicate_rows[['ragione_sociale', 'provincia', 'data_iscrizione_startup']].head())
else:
    print(f"‚úÖ Nessun duplicato trovato")

# 5. VALIDAZIONE DATI CHIAVE
print(f"\n‚úÖ VALIDAZIONE DATI CHIAVE:")

# Verifica province italiane
province_uniche = df_startup['provincia'].nunique()
print(f"Province uniche: {province_uniche}")

# Verifica settori
settori_unici = df_startup['settore'].nunique()
settori_list = df_startup['settore'].value_counts()
print(f"Settori unici: {settori_unici}")
print(f"Distribuzione settori:")
for settore, count in settori_list.items():
    if pd.notna(settore):
        print(f"  {settore}: {count}")

# Verifica classi produzione (dati REALI)
if 'classe_produzione' in df_startup.columns:
    prod_classes = df_startup['classe_produzione'].value_counts()
    print(f"\nClassi produzione (FATTURATO REALE):")
    for classe, count in prod_classes.items():
        if pd.notna(classe):
            print(f"  Classe {classe}: {count} startup")

# Verifica classi addetti (dati REALI)
if 'classe_addetti' in df_startup.columns:
    add_classes = df_startup['classe_addetti'].value_counts()
    print(f"\nClassi addetti (DIPENDENTI REALI):")
    for classe, count in add_classes.items():
        if pd.notna(classe):
            print(f"  Classe {classe}: {count} startup")


for col in ['ragione_sociale', 'provincia', 'settore']:
    empty_count = df_startup[col].astype(str).str.strip().eq('').sum()
    if empty_count > 0:
        print(f"‚ö†Ô∏è Colonna '{col}' contiene {empty_count} valori vuoti (stringhe)")
            
print("\nColonne con missing:")
print(missing_summary[missing_summary['Valori_Mancanti'] > 0].to_string(index=False))
print(f"\nüìä DATASET PULITO FINALE:")
print(f"Righe: {df_startup.shape[0]:,}")
print(f"Colonne: {df_startup.shape[1]}")
print(f"‚úÖ Colonne standardizzate")
print(f"‚úÖ Qualit√† verificata")
```

```python
# Script 5: Arricchimento Basato su Dati Reali (No Web Scraping)
print("=== ARRICCHIMENTO CON DATI REALI REGISTRO IMPRESE ===")
print("Sfruttando produzione, addetti e capitale REALI")
print("-" * 60)

# Non serve pi√π web scraping!

def enrich_with_real_data(df_startup):
    """
    Arricchisce il dataset utilizzando i dati REALI del Registro Imprese
    """
    print("üöÄ AVVIO ARRICCHIMENTO CON DATI REALI:")
    
    df_enriched = df_startup.copy()
    
    # 1. CALCOLA ET√Ä PRECISA AZIENDA
    print("1Ô∏è‚É£ Calcolo et√† azienda...")
    date_col = 'data_iscrizione_startup'
    df_enriched['company_age_years'] = (pd.Timestamp.now() - df_enriched[date_col]).dt.days / 365.25
    df_enriched['company_age_years'] = df_enriched['company_age_years'].round(1)
    
    # 2. DECODIFICA DATI REALI PRODUZIONE
    print("2Ô∏è‚É£ Decodifica produzione reale...")
    
    PRODUZIONE_VALUES = {
        'A': 50000,      # Punto medio 0-100k
        'B': 300000,     # Punto medio 100k-500k  
        'C': 750000,     # Punto medio 500k-1M
        'D': 1500000,    # Punto medio 1M-2M
        'E': 3500000,    # Punto medio 2M-5M
        'F': 7500000,    # Punto medio 5M-10M
        'G': 30000000,   # Punto medio 10M-50M
        'H': 75000000    # Stima >50M
    }
    
    df_enriched['estimated_revenue_eur'] = df_enriched['classe_produzione'].map(PRODUZIONE_VALUES)
    
    # Categorizza revenue per funding
    def categorize_revenue_for_funding(revenue):
        if pd.isna(revenue):
            return 'Non disponibile'
        elif revenue < 100000:
            return 'Early stage'
        elif revenue < 1000000:
            return 'Seed stage'
        elif revenue < 5000000:
            return 'Growth stage'
        else:
            return 'Scale-up'
    
    df_enriched['revenue_stage'] = df_enriched['estimated_revenue_eur'].apply(categorize_revenue_for_funding)
    
    # 3. DECODIFICA DATI REALI ADDETTI
    print("3Ô∏è‚É£ Decodifica addetti reali...")
    
    ADDETTI_VALUES = {
        'A': 2,    # Punto medio 0-4
        'B': 7,    # Punto medio 5-9
        'C': 15,   # Punto medio 10-19
        'D': 35,   # Punto medio 20-49
        'E': 150,  # Punto medio 50-249
        'F': 400   # Stima 250+
    }
    
    df_enriched['estimated_employees_num'] = df_enriched['classe_addetti'].map(ADDETTI_VALUES)
    
    # 4. DECODIFICA CAPITALE SOCIALE REALE
    print("4Ô∏è‚É£ Decodifica capitale sociale...")
    
    CAPITALE_VALUES = {
        '1': 1,                    # 1 ‚Ç¨
        '2': 2500,                 # 1-5k ‚Ç¨
        '3': 7500,                 # 5-10k ‚Ç¨
        '4': 30000,                # 10-50k ‚Ç¨
        '5': 75000,                # 50-100k ‚Ç¨
        '6': 175000,               # 100-250k ‚Ç¨
        '7': 375000,               # 250-500k ‚Ç¨
        '8': 750000,               # 500k-1M ‚Ç¨
        '9': 1750000,              # 1‚Äì2.5M ‚Ç¨
        '10': 3750000,             # 2.5‚Äì5M ‚Ç¨
        '11': 6000000              # 5M+ ‚Ç¨
    }
    
    df_enriched['estimated_capital_eur'] = df_enriched['classe_capitale'].astype(str).map(CAPITALE_VALUES)
    
    # 5. CALCOLA METRICHE BUSINESS REALI
    print("5Ô∏è‚É£ Calcolo metriche business...")
    
    # Revenue per employee (produttivit√†)
    df_enriched['revenue_per_employee'] = df_enriched['estimated_revenue_eur'] / df_enriched['estimated_employees_num']
    df_enriched['revenue_per_employee'] = df_enriched['revenue_per_employee'].replace([np.inf, -np.inf], np.nan)
    
    # Capital efficiency (revenue/capitale)
    df_enriched['capital_efficiency'] = df_enriched['estimated_revenue_eur'] / df_enriched['estimated_capital_eur']
    df_enriched['capital_efficiency'] = df_enriched['capital_efficiency'].replace([np.inf, -np.inf], np.nan)
    
    # Employee intensity (addetti/capitale)
    df_enriched['employee_capital_ratio'] = df_enriched['estimated_employees_num'] / (df_enriched['estimated_capital_eur'] / 1000)  # Per 1k di capitale
    df_enriched['employee_capital_ratio'] = df_enriched['employee_capital_ratio'].replace([np.inf, -np.inf], np.nan)
    
    # 6. STIMA FUNDING STAGE BASATO SU DATI REALI
    print("6Ô∏è‚É£ Stima funding stage da dati reali...")
    
    def estimate_funding_stage_real(row):
        age = row['company_age_years']
        revenue = row['estimated_revenue_eur']
        employees = row['estimated_employees_num']
        capital = row['estimated_capital_eur']
        
        # Logica basata su metriche reali
        if pd.isna(revenue) and pd.isna(employees):
            return 'Pre-seed' if age < 2 else 'Seed'
        
        # Basato principalmente su revenue (dato pi√π affidabile)
        if pd.notna(revenue):
            if revenue < 50000:
                return 'Pre-seed'
            elif revenue < 500000:
                return 'Seed'
            elif revenue < 2000000:
                return 'Series A'
            elif revenue < 10000000:
                return 'Series B'
            else:
                return 'Growth/Later'
        
        # Basato su dipendenti se revenue non disponibile
        elif pd.notna(employees):
            if employees <= 5:
                return 'Pre-seed' if age < 2 else 'Seed'
            elif employees <= 20:
                return 'Series A'
            elif employees <= 50:
                return 'Series B'
            else:
                return 'Growth/Later'
        
        # Fallback basato su et√†
        if age < 1:
            return 'Pre-seed'
        elif age < 3:
            return 'Seed'
        elif age < 5:
            return 'Series A'
        else:
            return 'Growth/Later'
    
    df_enriched['funding_stage_real'] = df_enriched.apply(estimate_funding_stage_real, axis=1)
    
    # 7. STIMA VALUTAZIONE AZIENDA
    print("7Ô∏è‚É£ Stima valutazione...")
    
    def estimate_valuation(row):
        revenue = row['estimated_revenue_eur']
        stage = row['funding_stage_real']
        sector = row['settore']
        is_tech = row.get('alto_valore_tech_energia', 'NO') == 'SI'
        
        if pd.isna(revenue):
            return None
        
        # Multipli di revenue per settore e stage
        base_multiple = 2.0  # Default
        
        if sector == 'SERVIZI':
            base_multiple = 3.0 if is_tech else 2.0
        elif sector == 'INDUSTRIA/ARTIGIANATO':
            base_multiple = 1.5
        elif sector == 'COMMERCIO':
            base_multiple = 1.0
        
        # Aggiustamento per stage
        stage_multiplier = {
            'Pre-seed': 0.5,
            'Seed': 1.0,
            'Series A': 1.5,
            'Series B': 2.0,
            'Growth/Later': 2.5
        }
        
        final_multiple = base_multiple * stage_multiplier.get(stage, 1.0)
        return revenue * final_multiple
    
    df_enriched['estimated_valuation_eur'] = df_enriched.apply(estimate_valuation, axis=1)
    
    # 8. ANALISI REQUISITI INNOVAZIONE
    print("8Ô∏è‚É£ Analisi requisiti innovazione...")
    
    # Conta requisiti soddisfatti
    req_cols = ['requisito_1_rd', 'requisito_2_team', 'requisito_3_brevetti']
    df_enriched['innovation_requirements_count'] = 0
    
    for col in req_cols:
        if col in df_enriched.columns:
            df_enriched['innovation_requirements_count'] += (df_enriched[col] == 'SI').astype(int)
    
    # Classifica livello innovazione
    def classify_innovation_level(count):
        if count == 0:
            return 'Tradizionale'
        elif count == 1:
            return 'Innovativa'
        elif count == 2:
            return 'Altamente Innovativa'
        else:
            return 'Super Innovativa'
    
    df_enriched['innovation_level'] = df_enriched['innovation_requirements_count'].apply(classify_innovation_level)
    
    # 9. SCORE COMPLESSIVO STARTUP
    print("9Ô∏è‚É£ Calcolo score complessivo...")
    
    def calculate_startup_score(row):
        score = 0
        
        # Score da produzione (0-8 punti)
        if pd.notna(row['estimated_revenue_eur']):
            if row['estimated_revenue_eur'] >= 10000000: score += 8
            elif row['estimated_revenue_eur'] >= 5000000: score += 7
            elif row['estimated_revenue_eur'] >= 2000000: score += 6
            elif row['estimated_revenue_eur'] >= 1000000: score += 5
            elif row['estimated_revenue_eur'] >= 500000: score += 4
            elif row['estimated_revenue_eur'] >= 100000: score += 3
            elif row['estimated_revenue_eur'] >= 50000: score += 2
            else: score += 1
        
        # Score da dipendenti (0-6 punti)
        if pd.notna(row['estimated_employees_num']):
            if row['estimated_employees_num'] >= 250: score += 6
            elif row['estimated_employees_num'] >= 50: score += 5
            elif row['estimated_employees_num'] >= 20: score += 4
            elif row['estimated_employees_num'] >= 10: score += 3
            elif row['estimated_employees_num'] >= 5: score += 2
            else: score += 1
        
        # Score da innovazione (0-6 punti)
        score += row['innovation_requirements_count'] * 2
        
        # Bonus caratteristiche speciali
        if row.get('alto_valore_tech_energia', 'NO') == 'SI': score += 2
        if row.get('vocazione_sociale', 'NO') == 'SI': score += 1
        if pd.notna(row.get('sito_web', '')): score += 1
        
        return score
    
    df_enriched['startup_score'] = df_enriched.apply(calculate_startup_score, axis=1)
    
    # Classifica tier
    def classify_tier(score):
        if score >= 20:
            return 'Tier S (Eccezionale)'
        elif score >= 15:
            return 'Tier A (Eccellente)'
        elif score >= 10:
            return 'Tier B (Ottima)'
        elif score >= 5:
            return 'Tier C (Buona)'
        else:
            return 'Tier D (Base)'
    
    df_enriched['startup_tier'] = df_enriched['startup_score'].apply(classify_tier)
    
    # 10. INVESTMENT READINESS
    print("üîü Valutazione investment readiness...")
    
    def assess_investment_readiness(row):
        score = row['startup_score']
        age = row['company_age_years']
        revenue = row['estimated_revenue_eur']
        
        if score >= 15 and age >= 1 and pd.notna(revenue) and revenue >= 100000:
            return 'Investment Ready'
        elif score >= 10 and age >= 0.5:
            return 'Nearly Ready'
        elif score >= 5:
            return 'Developing'
        else:
            return 'Early Stage'
    
    df_enriched['investment_readiness'] = df_enriched.apply(assess_investment_readiness, axis=1)
    
    # 11. TIMESTAMP E METADATA
    df_enriched['enrichment_date'] = pd.Timestamp.now().strftime('%Y-%m-%d')
    df_enriched['enrichment_method'] = 'real_registry_data'
    
    return df_enriched

# ESECUZIONE ARRICCHIMENTO
df_enriched = enrich_with_real_data(df_startup)

print(f"\n‚úÖ ARRICCHIMENTO COMPLETATO!")
print(f"Dataset originale: {len(df_startup)} startup, {len(df_startup.columns)} colonne")
print(f"Dataset arricchito: {len(df_enriched)} startup, {len(df_enriched.columns)} colonne")
print(f"Nuove colonne aggiunte: {len(df_enriched.columns) - len(df_startup.columns)}")

# Aggiorna dataset principale
df_startup = df_enriched

print(f"\nüéØ NUOVE COLONNE BASATE SU DATI REALI:")
new_columns = [
    'company_age_years', 'estimated_revenue_eur', 'revenue_stage',
    'estimated_employees_num', 'estimated_capital_eur', 'revenue_per_employee',
    'capital_efficiency', 'employee_capital_ratio', 'funding_stage_real',
    'estimated_valuation_eur', 'innovation_requirements_count', 'innovation_level',
    'startup_score', 'startup_tier', 'investment_readiness'
]

for col in new_columns:
    if col in df_startup.columns:
        print(f"  ‚úÖ {col}")

print(f"\nüíé VALORE AGGIUNTO:")
print(f"üìä Revenue reali stimate da Registro Imprese")
print(f"üë• Dipendenti reali da classificazione ufficiale")
print(f"üíº Capitale sociale da dati societari")
print(f"üìà Metriche di performance calcolate")
print(f"üéØ Funding stage basato su dati concreti")
print(f"üí∞ Valutazioni stimate con multipli di mercato")
print(f"üèÜ Score e tier basati su performance reali")
```

```python
# ============================================================================
# CELLA 2: FUNZIONE BENCHMARK (DA ESEGUIRE DOPO CELLA 1)
# ============================================================================

def create_sector_benchmarks(df_startup):
    """
    Crea benchmark completi per settore
    """
    try:
        print("üîÑ Creazione benchmark...")
        
        benchmark_metrics = df_startup.groupby('settore').agg({
            # Metriche di base
            'ragione_sociale': 'count',
            'company_age_years': ['mean', 'median'],
            
            # Metriche economiche
            'estimated_revenue_eur': ['mean', 'median', 'std'],
            'estimated_employees_num': ['mean', 'median'],
            'estimated_capital_eur': ['mean', 'median'],
            
            # Metriche di performance
            'revenue_per_employee': ['mean', 'median'],
            'capital_efficiency': ['mean', 'median'],
            'startup_score': ['mean', 'median'],
            
            # Metriche di innovazione
            'innovation_requirements_count': 'mean',
            
            # Percentuali
            'alto_valore_tech_energia': lambda x: (x == 'SI').mean() * 100 if not x.empty else 0,
            'vocazione_sociale': lambda x: (x == 'SI').mean() * 100 if not x.empty else 0,
        }).round(2)
        
        # Flatten column names correttamente
        benchmark_metrics.columns = ['_'.join(map(str, col)).strip() for col in benchmark_metrics.columns]
        benchmark_metrics['ranking_score'] = benchmark_metrics['startup_score_median'].rank(ascending=False) #colonna di ranking basato su score median
        benchmark_metrics['revenue_cv'] = benchmark_metrics['estimated_revenue_eur_std'] / benchmark_metrics['estimated_revenue_eur_mean']# collona du dispersione come coeficiente di variazione per revenue
        
        print("‚úÖ Benchmark creati!")
        print(f"üìä Settori analizzati: {len(benchmark_metrics)}")
        
        # Mostra prime colonne per debug
        print("\nüîç Prime 5 colonne create:")
        for i, col in enumerate(benchmark_metrics.columns[:5]):
            print(f"  {i+1}. {col}")
            
        return benchmark_metrics
        
    except Exception as e:
        print(f"‚ùå Errore in create_sector_benchmarks: {e}")
        return pd.DataFrame()

# Esegui la funzione
benchmarks = create_sector_benchmarks(df_startup)
```
